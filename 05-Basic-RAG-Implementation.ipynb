```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
loader = TextLoader("data/state_of_the_union.txt")

loaded_document = loader.load()

loaded_document
```

```python
print(loaded_document[0].page_content)
```

```python
text_splitter = CharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=0
)

chunks_of_texts = text_splitter.split_documents(loaded_document)

chunks_of_texts
```

```python
len(chunks_of_texts)
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")

embeddings
```

```python
embeddings.embed_query("Hello, world")
```

```python
vector_store = FAISS.from_documents(
    chunks_of_texts,
    embeddings
)
```

```python
vector_store
```

```python
retriever = vector_store.as_retriever()  ## Default search_kwargs=4
```

```python
retriever.invoke("what did he say about ketanji brown jackson?")
```

### Simple use With LCEL (LangChain Expression Language)

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
template = """Answer the question based only on the following context:

{context}

Question: {question}
"""
```

```python
prompt_template = ChatPromptTemplate.from_template(template=template)

prompt_template
```

```python
llm = ChatGroq(model="llama-3.3-70b-versatile")

llm
```

```python
def format_docs(docs):
    return "\n\n".join([d.page_content for d in docs])
```

```python
chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt_template
    | llm
    | StrOutputParser()
)
```

```python
response = chain.invoke("what did he say about ketanji brown jackson?")

response
```
